```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import pickle
from imblearn.over_sampling import SMOTE
```

```python
#load the data we got from the preprocessing

df = pd.read_csv('./churndata(1).csv')
df
```

Drop the Unnamed: 0 column

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df
```

For the next step, we set the X and Y variables (X will represent all the independent variables whereas Y will be for the dependent variable which is Churn)

```python
#setting the x variable
x = df.drop("Churn", axis=1)
x
```

```python
#setting the y variable
y =  df['Churn']
y
```

Once the parameters are set, we classify the data into x_train,y_train,x_test and y_test segments

In our case, we'll use 20% for test and 80% for training

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2)
```

# DECISION TREE CLASSIFICATION

There are many ways in which we can do our classification and in this particular use case, decision tree and random forest classification techniques will be applied

```python
model = DecisionTreeClassifier(criterion='gini', random_state=100, max_depth=6, min_samples_leaf=8)
```

Once the decision tree classifier is called, the train data needs to be fitted

```python
model.fit(x_train,y_train)
```

we then create a y prediction and pass an x_test for it

```python
y_pred = model.predict(x_test)
y_pred
```

A comparison is done between the predicted and actual y values and from here we can get an accuracy score for our classification

```python
print(classification_report(y_test, y_pred, labels=[0,1]))
```

```python
#get the confusion matrix
print(confusion_matrix(y_test,y_pred))
```

From the report on the model score, we can see that we have a precision score of 58%, recall score of 44% and f1-score of 50% for the churn prediction. These very low scores occur because the dataset is imbalanced. A fine tuning of the models can be done to improve these scores.

For Fine tuning, SMOTEENN will be used

SMOTTEEN is used on imbalanced data to combine oversampling and undersampling using SMOTE and Editen Nearest Neighbours(ENN)


```python
sm = SMOTE()
X_resampled, y_resampled = sm.fit_resample(x,y)
```

```python
xr_train,xr_test,yr_train,yr_test = train_test_split(X_resampled,y_resampled,test_size=0.2)
```

```python
smote_model = DecisionTreeClassifier(criterion='gini', random_state=100, max_depth=6, min_samples_leaf=8)
```

```python
smote_model.fit(xr_train,yr_train)
```

```python
smote_y_pred = smote_model.predict(xr_test)
```

```python
print(classification_report(yr_test, smote_y_pred, labels=[0,1]))
```

```python
print(confusion_matrix(yr_test,smote_y_pred))
```

# RANDOM FOREST CLASSIFICATION

Initialization of the model using RandomForestClassifier imported from sklearn

```python
rf_model = RandomForestClassifier(n_estimators=100,criterion='gini', random_state=100, max_depth=6, min_samples_leaf=8)
```

```python
rf_model.fit(x_train,y_train)
```

```python
y_pred_rf = rf_model.predict(x_test)
```

```python
print(classification_report(y_test, y_pred_rf, labels=[0,1]))
```

From random forest classification, the model precision at 67%, 37% recall and 48% f1-score. These are low accuracy values.

A fine tuning of the random forest classification model can again be done using SMOTEENN

```python
sm = SMOTE()
x_resampled, y_resampled = sm.fit_resample(x,y)
```

```python
xr_train,xr_test,yr_train,yr_test = train_test_split(x_resampled,y_resampled,test_size=0.2)
```

```python
smote_model_rf = RandomForestClassifier(criterion="gini", random_state=100, max_depth=6, min_samples_leaf=8)
```

```python
smote_model_rf.fit(xr_train,yr_train)
```

```python
smote_y_pred_rf = smote_model_rf.predict(xr_test)
```

```python
print(classification_report(yr_test,smote_y_pred_rf, labels=[0,1]))
```

```python
print(confusion_matrix(yr_test, smote_y_pred_rf))
```

The two models provide us with nearly the same accuracy values in prediction. We then need to save our models because they'll need to be used later in case we might need to make a prediction. Pickle will be used for this task

# SAVING THE MODEL

```python
filename = "smotechurnmodel.sav"
```

```python
pickle.dump(smote_model_rf, open(filename, 'wb'))
```

```python
load_model = pickle.load(open(filename, 'rb'))
```

```python
load_model.score(x_test,y_test)
```

Our model is able to predict customer churn with an accuracy of 77%.
